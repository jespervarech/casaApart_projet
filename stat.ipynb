```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
apart = pd.read_csv("apart_sans_na.csv")
```

```python
apart
```

```python
apart.isnull().sum()
```

```python
apart.describe()
```

```python
# Boxplot des prix par secteur
plt.figure(figsize=(20, 8))
sns.boxplot(data=apart, x='secteur', y='prix')
plt.title('Répartition des prix par secteur')
plt.xticks(rotation=45)
plt.savefig('Repartition_des_prix_par_secteur.png', dpi=300)
plt.show()
```
